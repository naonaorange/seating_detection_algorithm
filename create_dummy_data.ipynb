```python
%matplotlib inline

import numpy as np
import matplotlib.pylab as plt

x = np.linspace(0, np.pi, 100)
y = np.sin(x)
y =  y + np.random.rand(100) / 10

z = np.concatenate([np.array([1.], float), y], 0)

LOG_PATH = "dummy_1.csv"
with open(LOG_PATH, mode='w') as f:
    try:
        for a in range(1000):
            x = np.linspace(0, np.pi, 100)
            y = np.sin(x)
            y =  y + np.random.rand(100) / 10

            z = np.concatenate([np.array([1.], float), y], 0)
            
            count = 0
            for i in z:
                f.write(str(i))
                if count != 100:
                    f.write(',')
                count = count + 1
            f.write('\n')

    except Exception as e:
        print(e.message)
        pass

plt.plot(x, y)
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.show()
```
